Agregámos três tabelas escolhidas: consumos mensais por concelho, ordens de serviço e número de contratos de energia.

Seguimos então para a **análise e correlação entre as variáveis**. As variáveis presentes no nosso dataset são as seguintes:

**Ano e Mês:**
Ano e mês indicam o **período temporal** em que os dados foram registados. Podemos analisar as tendências ao longo do ano e identificar variações sazonais nos serviços realizados.

**Concelho:**
O concelho representa a **área geográfica** onde os serviços foram realizados. Podemos analisar a distribuição geográfica dos serviços e identificar se há variações nos tipos de serviço entre diferentes concelhos.

**Nível de Tensão:**
O nível de tensão indica o tipo de tensão da instalação elétrica onde os serviços foram realizados. Podem ser do tipo Muito Alta, Alta e Média Tensões ou Média tensão.

**Energia Ativa (kWh):**
A energia ativa representa a quantidade de energia elétrica consumida.

**Tipo de Serviço:**
O tipo de serviço indica a natureza do trabalho realizado. Podem ser do tipo Serviço de Corte e Re-Ligação.

**Ordens de Serviço Realizadas (#):**
Este é o número de ordens de serviço realizadas.

**Inclui contador inteligente:**
Indica se o serviço realizado inclui a utilização de um contador inteligente.

**Número de CPE's:**
Representa o número de Contratos de Pontos de Entrega associados aos serviços realizados.

Podemos ainda **verificar a correlação entre as variáveis:**

**Concelho e Nível de Tensão:**
Podemos analisar se existem diferenças nos níveis de tensão entre os concelhos. Por exemplo, podemos verificar se certos concelhos têm predominantemente uma tensão específica.

**Energia Ativa e Número de CPE's:**
Podemos investigar se há uma correlação entre a quantidade de energia ativa consumida e o número de clientes afetados. Isso pode ajudar a entender se áreas com maior consumo de energia têm mais frequentemente serviços de corte e re-ligação.

**Tipo de Serviço e Contador Inteligente:**
Podemos examinar se a presença de um contador inteligente afeta o tipo de serviço realizado. Por exemplo, pode ser que a presença de contadores inteligentes esteja associada a uma eficiência operacional maior em certos tipos de serviços.

**Concelho e Número de CPE's:**
Podemos investigar se há variações no número de clientes de pontos de entrega entre os concelhos. Isso pode indicar diferenças na densidade populacional ou no tamanho dos mercados em cada área.

**Tipo de Serviço e Energia Ativa:**
Podemos explorar se há alguma relação entre o tipo de serviço realizado e a quantidade de energia ativa envolvida. Por exemplo, se certos tipos de serviços estão associados a maiores consumos de energia.

**Análise de outliers**

Os outliers encontrados nestas bases de dados são dados que remetem para zonas mais populosas por isso não são valores fora do esperado.

```python
def find_outliers(f):
    q1 = f.quantile(0.25)
    q3 = f.quantile(0.75)
    iqr = q3 - q1
    return f[(f < (q1 - 1.5 * iqr)) | (f > (q3 + 1.5 * iqr))]
```

```python
find_outliers(consumosMensaisPorConcelho_df["Energia Ativa (kWh)"])
```



**Criação de dados totalmente categóricos**

Para a criação de dados totalmente categóricos foi usado o método de discretização.

Energia Ativa (kWh) - Foram definidos intervalos de [0,23000000] para "Baixa", [23000001,46000000] para "Média" e [46000001,69000000] para "Alta". Escolhemos estes intervalos após analisarmos os valores presentes na base de dados.

Número de CPE's - Foram definidos intervalos de [0,15000] para "Baixa", [15001,30000] para "Média" e [30001,59000] para "Alta". Escolhemos estes intervalos após analisarmos os valores presentes na base de dados.

Ordens Serviço Realizadas (#) - Foram defininidos intervalos de [0,10] para "Inferior a 10" e [11,100] para "Superior a 10". Escolhemos estes intervalos após analisarmos os valores presentes na base de dados.


```python
#Usar discretização para transformar os dados em totalmente categóricos
category_df= merged_df.copy()
category_df['Energia Ativa (kWh)'] = pd.cut(category_df['Energia Ativa (kWh)'],
                                                      bins=[0,23000000,46000000,69000000],
                                                      labels=['Baixa', 'Média', 'Alta'])

category_df["Número de CPE's"] = pd.cut(category_df["Número de CPE's"],
                                                      bins=[0,15000,30000,50000],
                                                      labels=['Baixa', 'Média', 'Alta'])


category_df['Ordens Serviço Realizadas (#)'] = pd.cut(category_df['Ordens Serviço Realizadas (#)'],
                                                      bins=[0,10,100],
                                                      labels=['Inferior a 10', 'Superior a 10'])

category_df.drop_duplicates(inplace=True)
category_df.info()
display(category_df)
```

## **Normalização**
Após a criação das duas tabelas, numérica e categórica, podemos fazer a normalização dos dados.
Das diferentes técnicas existentes foram testadas duas: min-max e z-score. Contudo a que será utilizada nos seguintes passos é a min-max, visto que dos resultados obtidos, estes foram os que melhor se adequaram aos valores desejados.

**Normalização Min-Max**

**Cluster Hierárquico**

O clustering hierárquico é uma técnica de agrupamento que constrói clusters aninhados, unindo-os ou dividindo-os sucessivamente. Como anteriormente já foi calculado o número de cluesters ideais (n = 4), podemos apenas fazer a construção das folhas necessárias.

Criámos uma amostra com cerca de 2% dos dados do nosso modelo devido à limitação de processamento do colab. Faremos a análise do dendograma com base nesta amostra.

```python

amostra = pca_df.sample(frac=0.02, random_state=42)

hierarchical_k = AgglomerativeClustering(n_clusters=4).fit(amostra)
sns.scatterplot(amostra, x='PC1', y='PC2', hue=hierarchical_k.labels_).set(title='Hierarchical Clustering')
plt.show()
```

Depois da realização dos clusters hieráquicos, para conseguir uma avaliação do resultado obtido, é necessário a construção do dendograma para visualizar os clusters construídos.

Para isso, criamos a função dendogram que de forma recursiva cria visualmente as folhas do algoritmo.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
def plot_dendrogram(model, **kwargs):
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

```python
hierarchical_full = AgglomerativeClustering(n_clusters=None, distance_threshold=0).fit(amostra)
plot_dendrogram(hierarchical_full)
plt.title('Dendrograma')
plt.xlabel('Índices dos pontos')
plt.ylabel('Distância')

num_clusters = 4

Z = linkage(amostra, 'ward')
plt.axhline(y=Z[-num_clusters+1, 2], color='r', linestyle='--')
plt.show()
```

Ao analisar o dendrograma podemos concluir que existem 3 grandes grupos de dados.
Através do método do número de clusters definimos 4 clusters presentes no dendograma.
A distância entre os pontos dentro dos clusters não é superior a 0,5.
